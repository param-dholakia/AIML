```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Housing.csv")
df
```

```python
df.columns
```

```python
X=df['area']
```

```python
y=df['price']
```

```python
y_mean=y.mean()
print("Mean of target column: ",y_mean)
```

```python
X_mean=X.mean()
print("Mean of Feature Column(Area mean): ",X_mean)
```

```python
n=df.shape[0]
print("No. of rows: ",n)
```

```python
plt.scatter(X,y)
plt.show()
```

```python
nuemerator_sum=0
denominator_sum=0
m=0
for i in range(n):
    x1=X[i]-X_mean
    y1=y[i]-y_mean
    nuemerator_sum+=((x1)*(y1))
    denominator_sum+=x1*x1

print("Neumerator: ",nuemerator_sum)
print("Denominator: ",denominator_sum)
m=nuemerator_sum/denominator_sum
print("m: ",m)
```

```python
b=y_mean-(m*X_mean)
print("b: ",b)
```

### Prediction starts here...

```python
area=input("Enter Area: ")
price=(m*float(area))+b
print("Price: ",round(price,-2))
```
