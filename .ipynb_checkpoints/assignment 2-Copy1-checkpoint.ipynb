# Part 2: Preprocessing

Standard Scaling.

```python
df1.shape[0]
```

```python
df1.isna().sum()
```

```python
bedroom_mode=df1['total_bedrooms'].mode()
```

```python
print(bedroom_mode)
bedroom_mode=bedroom_mode[0]
```

```python
df1.fillna({'total_bedrooms':bedroom_mode}, inplace=True)  # Fills missing values in a specific column
```

```python
df1.isna().sum()
```

```python
df1.head(10) 
```

```python
### Normalizing my dataset...
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
#created objects
min_max=MinMaxScaler()
ss=StandardScaler()
```

```python
y1=df1['median_house_value']
X1 = df1.drop(columns=['median_house_value'], axis=1)
```

```python
#applying Min-Max Scaling
X1_scaled=ss.fit_transform(X1)
X1 = pd.DataFrame(X1_scaled, columns=X1.columns)

print(X1.head(3))
```

```python
X1.columns
```

```python
#linear graphs
import matplotlib.pyplot as plt
j=1
plt.figure(figsize=(50,5))
for i in X1.columns:
    plt.subplot(1,9,j)
    plt.scatter(X1[i],y1)
    j+=1

plt.show()
```

```python
#splitting the dataset
from sklearn.model_selection import train_test_split
```

```python
X1.columns
```

```python
X1_train,X1_test,y1_train,y1_test=train_test_split(X1,y1,test_size=0.2,random_state=42)
```

```python
#model importing and training
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
lr=LinearRegression()
```

```python
lr.fit(X1_train,y1_train)
```

```python
y1_pred=lr.predict(X1_test)
```

```python
y1_pred=y1_pred.flatten()
y1_test=y1_test.to_numpy().flatten()
```

```python
data1=pd.DataFrame({'Actual':y1_test,'Predicted': y1_pred})
data1
```

```python
mse = mean_squared_error(y1_test, y1_pred)
rmse = mse ** 0.5
rmse
```

```python
r2 = r2_score(y1_test, y1_pred)
r2*100
```

```python
mae=mean_absolute_error(y1_pred,y1_test)
mae
```
