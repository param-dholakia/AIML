## Part 5: Experimentation
1. Train at least one additional regression model (e.g., Decision Tree Regressor or Random
Forest Regressor).
2. Compare the performance of both models using the evaluation metrics. 

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
# Train Decision Tree Regressor
y1_train=y1_train.to_numpy().flatten()

dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X1_train, y1_train)

# Train Random Forest Regressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X1_train, y1_train)
```

```python
# Predictions
y1_pred_dt = dt_model.predict(X1_test)  # Decision Tree
y1_pred_rf = rf_model.predict(X1_test)  # Random Forest
```

```python
# Evaluation Metrics
models = {
    "Linear Regression": y1_pred_lr,
    "Decision Tree": y1_pred_dt,
    "Random Forest": y1_pred_rf
}

for model_name, y_pred in models.items():
    rmse=np.sqrt(mean_squared_error(y1_test, y_pred))
    r2=r2_score(y1_test, y_pred)
    mae=mean_absolute_error(y1_test,y_pred)
    print(model_name," - RMSE: ",rmse," R2 Score: ",r2*100," MAE: ",mae)
```
